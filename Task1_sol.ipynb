```python
#Data summary for the simulation
def dist_summary(dist, names = 'dist_name'):
    """Plot a frequency distribution and return a summary of it"""
    import pandas as pd
    import matplotlib.pyplot as plt
    ser = pd.Series(dist)
    fig = plt.figure(1, figsize=(9, 6))
    ax = fig.gca()
    ser.hist(ax = ax, bins = 120)
    ax.set_title('Frequency distribution of ' + names)
    ax.set_ylabel('Frequency')
    plt.show()
    return(ser.describe())
```

```python
#Method for the poisson distribution simulation for MODEL 1
distA = []
distB = []
def sim_poisson(nums, mean1 = manUOGoals.mean(), mean2 = manCOGoals.mean()):
    """ Simulates the Poisson distribution and gives its summary & plot"""
    import numpy as np
    import numpy.random as nr
    for n in nums:
        dist1 = nr.poisson(lam = mean1, size = n)
        dist2 = nr.poisson(lam = mean2, size = n)
        distA = list(dist1)
        distB = list(dist2)
        titl = 'Poisson distribution with ' + str(n) + ' values'
        print(dist_summary(dist1, titl))
        print(dist_summary(dist2, titl))        
        print('Emperical 95% CIs')
        print(np.percentile(dist1, [2.5, 97.5]))
        print(np.percentile(dist2, [2.5, 97.5]))        
        print(' ')
    return(dist1,dist2)
```

```python
#Run the simulation
distA,distB = sim_poisson([1000000])
```

```python
#method to calculate who wins
def winPercent(manU,manC):
    winManU = 0
    winManC = 0    
    draw = 0
    for x,y in zip(manU,manC):
        if x > y:
            winManU +=1
        if y > x:
            winManC +=1
        if x == y:
            draw +=1
    totalGames = winManU+winManC+draw
    manUWinPercent = winManU/totalGames * 100
    manCWinPercent = winManC/totalGames * 100
    gameDraw = draw/totalGames * 100
    print("manU wins: "+str(float("{0:.2f}".format(manUWinPercent)))+"%")
    print("manC wins: "+str(float("{0:.2f}".format(manCWinPercent)))+"%")
    print("Draw: "+str(float("{0:.2f}".format(gameDraw)))+"%")
```

```python
#calculate the win percentage. MODEL 1
winPercent(distA,distB)
```
